```python
#question no. 4
import json
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]
with open(filename) as f:
    data = json.load(f)

total_records = len(data)
missing_first_name = sum(1 for record in data if not record['patientDetails'].get('firstName'))
missing_last_name = sum(1 for record in data if not record['patientDetails'].get('lastName'))
missing_dob = sum(1 for record in data if not record['patientDetails'].get('birthDate'))

percentage_missing_first_name = (missing_first_name / total_records) * 100
percentage_missing_last_name = (missing_last_name / total_records) * 100
percentage_missing_dob = (missing_dob / total_records) * 100

print(f"Percentage of missing firstName: {percentage_missing_first_name:.2f}%")
print(f"Percentage of missing lastName: {percentage_missing_last_name:.2f}%")
print(f"Percentage of missing DOB: {percentage_missing_dob:.2f}%")
```

```python
#question no. 5
import json
from collections import Counter
from google.colab import files

uploaded = files.upload()

filename = list(uploaded.keys())[0]
with open(filename) as f:
    data = json.load(f)

genders = [record['patientDetails'].get('gender') for record in data if record['patientDetails'].get('gender')]
gender_mode = Counter(genders).most_common(1)[0][0]

for record in data:
    if not record['patientDetails'].get('gender'):
        record['patientDetails']['gender'] = gender_mode

total_patients = len(data)
female_count = sum(1 for record in data if record['patientDetails']['gender'].lower() == 'female')
percentage_female = (female_count / total_patients) * 100

print(f"Percentage of female patients after imputation: {percentage_female:.2f}%")
```

```python
#question no.6
import json
from datetime import datetime
from google.colab import files

uploaded = files.upload()

filename = list(uploaded.keys())[0]
with open(filename) as f:
    data = json.load(f)

def calculate_age(birth_date):
    if birth_date:
        birth_date = datetime.strptime(birth_date, '%Y-%m-%dT%H:%M:%S.%fZ')
        today = datetime.today()
        return today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return None

def categorize_age(age):
    if age is not None:
        if age <= 12:
            return 'Child'
        elif 13 <= age <= 19:
            return 'Teen'
        elif 20 <= age <= 59:
            return 'Adult'
        elif age >= 60:
            return 'Senior'
    return None

adult_count = 0
for record in data:
    birth_date = record['patientDetails'].get('birthDate')
    age = calculate_age(birth_date)
    age_group = categorize_age(age)
    record['ageGroup'] = age_group
    if age_group == 'Adult':
        adult_count += 1

print(f"Count of Adult patients: {adult_count}")
```

```python
#question no.8
import json
from collections import Counter
from google.colab import files

uploaded = files.upload()

filename = list(uploaded.keys())[0]
with open(filename) as f:
    data = json.load(f)

medicine_counter = Counter()

for record in data:
    medicines = record['consultationData']['medicines']
    for medicine in medicines:
        medicine_counter[medicine['medicineName']] += 1

most_common_medicines = medicine_counter.most_common(3)
third_most_frequent_medicine = most_common_medicines[2][0] if len(most_common_medicines) >= 3 else None

print(third_most_frequent_medicine)
```

```python
#question no. 9
import json
from google.colab import files

uploaded = files.upload()

filename = list(uploaded.keys())[0]
with open(filename) as f:
    data = json.load(f)

total_active_medicines = 0
total_inactive_medicines = 0

for record in data:
    medicines = record['consultationData']['medicines']
    for medicine in medicines:
        if medicine['isActive']:
            total_active_medicines += 1
        else:
            total_inactive_medicines += 1

total_medicines = total_active_medicines + total_inactive_medicines

percentage_active = (total_active_medicines / total_medicines) * 100 if total_medicines > 0 else 0
percentage_inactive = (total_inactive_medicines / total_medicines) * 100 if total_medicines > 0 else 0

print(f"Active Medicines: {percentage_active:.2f}%")
print(f"Inactive Medicines: {percentage_inactive:.2f}%")
```

```python
# question no. 11
import json
import numpy as np
from datetime import datetime
from google.colab import files

# Upload the file
uploaded = files.upload()

# Load the dataset
filename = list(uploaded.keys())[0]
with open(filename) as f:
    data = json.load(f)

# Function to calculate age from birthdate
def calculate_age(birth_date):
    if birth_date:
        birth_date = datetime.strptime(birth_date, '%Y-%m-%dT%H:%M:%S.%fZ')
        today = datetime.today()
        return today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return None

# Prepare the data for correlation calculation
ages = []
num_medicines = []

for record in data:
    birth_date = record['patientDetails'].get('birthDate')
    age = calculate_age(birth_date)
    if age is not None:
        num_meds = len(record['consultationData']['medicines'])
        ages.append(age)
        num_medicines.append(num_meds)

# Calculate the Pearson correlation
if len(ages) > 1 and len(num_medicines) > 1:
    correlation = np.corrcoef(ages, num_medicines)[0, 1]
else:
    correlation = 0

print(round(correlation, 2))
```
